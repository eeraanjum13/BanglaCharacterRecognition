```python
import pandas as pd
```

```python
df = pd.read_csv('train_labels2.csv')
```

```python
df
```

```python
df['Letter']
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
new_df = df.drop(['Id'] , axis = 1)
```

```python
y = new_df['Letter']
new_df = new_df.drop(['Letter'] , axis = 1)
```

```python
X = new_df
```

```python
clf = MLPClassifier()

clf.fit(X, y)
```

```python
clf.score(X,y)
```

```python
print(clf.predict(X))
```

```python
from sklearn.linear_model import Perceptron

plf = Perceptron()
plf.fit(X, y)

plf.score(X,y)
```

```python
print(plf.predict(X))
```
